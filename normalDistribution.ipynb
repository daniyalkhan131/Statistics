```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
#generate 100000 normal random integers with specified mean and std.
draw=norm.ppf(np.random.random(100000), loc=1000, scale=200).astype(int) #loc=mean and scale=std.
sns.distplot(draw)
```

```python
z=list()
for i in draw:
  z.append((i-1000)/200)
```

```python
a=np.array(z)
```

```python
print(a.mean())
print(np.round(a.mean()))
```

```python
print(a.std())
print(np.round(a.std()))
```
