```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10  # number of trials
p = 0.8  # probability of success
size = 1000  # number of samples to generate

#here we are conducting task of tossing 10 coins 1000 times
#and by this pdf is made or plotted
binomial_dist = np.random.binomial(n, p, size)

print(binomial_dist.shape)
binomial_dist[:15] #it will give array of 1000 size telling out of 10 coins how
#much head appears(we are considering head as success)
#as p=0.8 then it gives head more
```

```python
p = 0.5
binomial_dist = np.random.binomial(n, p, size)

plt.hist(binomial_dist,density=True)
plt.show()
```

```python
p = 0.1
binomial_dist = np.random.binomial(n, p, size)

plt.hist(binomial_dist,density=True)
plt.show()
```
