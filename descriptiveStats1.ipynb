```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('train_titanic.csv')
```

```python
df.sample(5)
```

```python
#Pclass is categorical data
#bar chat
df['Pclass'].value_counts().plot(kind='bar')
```

```python
#pie chart
df['Pclass'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
#cumulative frequency table
# we change count to percent so that last category cumulative freq will be 100

a=df['Pclass'].value_counts()
a
```

```python
a=100*(a/a.sum())
```

```python
a
```

```python
b=a.values
b
```

```python
c=[]
cf=0
for i in b:
  cf=cf+i
  c.append(cf)
c
```

```python
plt.plot(c,marker='D')
```
