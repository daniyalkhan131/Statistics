```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
temp = sorted(df['sepal_length'].tolist())
```

```python
y_quant = []

for i in range(1,101):
  y_quant.append(np.percentile(temp,i)) #for calculatinf percentile of i in temp data
```

```python
samples = np.random.normal(loc=0,scale=1,size=1000)
```

```python
x_quant = []

for i in range(1,101):
  x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

```python
# using statsmodel

import statsmodels.api as sm
import matplotlib.pyplot as plt

# QQ plot of the two sets of data
fig = sm.qqplot(df['sepal_length'], line='45', fit=True) #default is normal dist.-
#-so it has default set as comparing to nrmal dist


plt.title('QQ Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

```python
#for uniform distribution
import scipy.stats as stats

# Generate a set of random data
x = np.random.uniform(low=0, high=1, size=1000)
```

```python
plt.hist(x)
```

```python
# Fit a uniform distribution to the data
params = stats.uniform.fit(x)
dist = stats.uniform(loc=params[0], scale=params[1])
```

```python
# Create a QQ plot of the data using the uniform distribution
fig = sm.qqplot(x, dist=dist, line='45')

plt.title('QQ Plot of Uniform Distribution with Uniform Fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

plt.show()
```
