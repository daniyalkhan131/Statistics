```python
import pandas as pd
import seaborn as sns
```

```python
titanic=pd.read_csv('train_titanic.csv')
```

```python
sns.kdeplot(titanic['Age'])
```

```python
titanic['Age'].mean()
```

```python
x = (titanic['Age'] - titanic['Age'].mean())/titanic['Age'].std()
```

```python
#can consider this to be normal distributed
titanic['Age'].skew()
```

```python
r1=titanic['Age'].mean() + 3*titanic['Age'].std()
r2=titanic['Age'].mean() - 3*titanic['Age'].std()
print(r1,r2)
```

```python
#outliers
titanic[(titanic['Age'] > r1) | (titanic['Age']<r2)]
```
